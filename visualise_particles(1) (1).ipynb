# Dynamics of Non-Interacting Particles

This notebook will read in and visualise the trajectory generated by the C++ code. 

```python
# function to read in the trajectory generated from the C++ code
def read_trajectory(filename):

    data = open(filename)
    file_contents = data.readlines()
    
    nsteps = int(file_contents[0].split()[0])
    N = int(file_contents[0].split()[1])
    L = int(file_contents[0].split()[2])
    
    positions = []
    l = 1
    for istep in range(nsteps):
        positions_this_step = []
        for i in range(N):
            positions_this_step.append([float(file_contents[l].split()[0]), float(file_contents[l].split()[1])])
            l += 1
        positions.append(positions_this_step)

    return positions, L

all_positions, L = read_trajectory("particles.txt")
```

```python
# function to display a snapshot of particles
def plot_snapshot(ax, positions):
    ax.plot([pos[0] for pos in positions], [pos[1] for pos in positions], marker='o', ms=10, ls='', color='k')
    fig.canvas.draw()
    
import matplotlib.pyplot as plt
%matplotlib notebook

fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])

dL = 10

# display the positions
for p in range(0, len(all_positions)):
    ax.clear()
    # draw the ground and walls
    ax.hlines(0, 0, L, ls='-', lw=3, color='k')
    ax.vlines(0, 0, L+dL, ls='-', lw=3, color='k')
    ax.vlines(L, 0, L+dL, ls='-', lw=3, color='k')

    ax.set_xlim(-dL, L+dL)
    ax.set_ylim(-dL, L+dL)
    plot_snapshot(ax, all_positions[p])
```
